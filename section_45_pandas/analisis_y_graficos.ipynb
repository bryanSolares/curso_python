```python
import pandas as pd
import cufflinks as cf
from IPython.display import display, HTML

cf.set_config_file(sharing='public', theme='space', offline=True)
cf.getThemes()
```

```python
data = pd.read_csv("./Referencia/pais.csv")
data
```

```python
data = data.dropna()
data
```

```python
# Pivotear una tabla
data = data.pivot(index='Year', columns='Pais', values='Poblacion')
data
```

```python
# GrÃ¡fico de lÃ­neas
data[['Albania','Algeria','India', 'China']].iplot(kind="line", xTitle="Year", yTitle="Population", title="GrÃ¡fico")
```

```python
# GrÃ¡fico de barras
data[['Albania','Algeria','India', 'China']].iplot(kind="bar", xTitle="Year", yTitle="Population", title="GrÃ¡fico")
```

```python
data[data.index.isin([1990])][['Albania','Algeria','India', 'China', "Brazil", "United States"]].iplot(kind="bar", xTitle="Year", yTitle="Population", title="GrÃ¡fico")
```

```python
# Modificando el orden de los grÃ¡ficos
data[data.index.isin([2018,2019,2020])][['Albania','Algeria','India', 'China', "Brazil", "United States"]].T.iplot(kind="bar", xTitle="Year", yTitle="Population", title="GrÃ¡fico")
```

```python
# Nuevo tipo de grÃ¡fico TIPO CAJA
data[['Albania','Algeria','India', 'China', "Brazil", "United States"]].iplot(kind="box")
```

```python
# Nuevo tipo de grÃ¡fico TIPO HISTOGRAMA
data[['Albania','Algeria','India', 'China', "Brazil", "United States"]].iplot(kind="hist")
```

```python
# Nuevo tipo de grÃ¡fico TIPO PASTEL
data2020 = data[data.index.isin([2020])][['Albania','Algeria','India', 'China', "Brazil", "United States"]].T

data2020 = data2020.rename(columns={2020:"2020"})

data2020.reset_index().iplot(kind="pie", labels="Pais", values="2020")
```
