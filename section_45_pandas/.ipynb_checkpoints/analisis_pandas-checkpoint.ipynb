```python
# Comparación de acuerdo a un dato estadístico
data_players = data[["short_name", "age", "height_cm", "nationality", "power_strength", "skill_dribbling", "club","overall", "value_eur"]]
real_madrid = data_players[data_players["club"]=="Real Madrid"]
barcelona = data_players[data_players["club"]=="FC Barcelona"]
juventus = data_players[data_players["club"]=="Juventus"]

plt.figure()
plt.subplot(121)
plt.boxplot(real_madrid["overall"])
plt.ylabel("Overall")
plt.title("Real Madrid")
plt.grid(True)

```

```python
plt.subplot(121)
plt.boxplot(barcelona["overall"])
plt.ylabel("Overall")
plt.title("Barcelona")
plt.grid(True)
```

```python
plt.subplot(121)
plt.boxplot(juventus["overall"])
plt.ylabel("Overall")
plt.title("Juventus")
plt.grid(True)
```
